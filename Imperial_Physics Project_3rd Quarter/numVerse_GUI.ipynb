```python
def showMenu():
    def on_closing():
        global root_menu,menuTracker
        #if tk.messagebox.askokcancel("Quit", "Close menu?"):
        menuTracker = 0
        root_menu.destroy()

    global root_menu,menuTracker
    boolHere = menuTracker%2
    if boolHere ==  0:
        menuTracker += 1
        root_menu = tk.Tk()
        root_menu.geometry("150x160+400+200")
        root_menu.title('MENU')
        root_menu.maxsize( width = 150, height = 160 )
        root_menu.minsize( width = 150, height = 160 )
        root_menu.resizable(0,0)

        canvas_menu_bar = tk.Canvas(root_menu, width = 150, height = 35)
        canvas_menu_bar.config(bg = black, bd = 0 )
        canvas_menu_bar.pack()

        #adds the "Let's Explore"- LABEL
        label = tk.Label(canvas_menu_bar,text = "M E N U  Ξ", font = 'none 16 bold', bg = black, fg = white , bd = 0) 
        canvas_menu_bar.create_window(75, 20, window = label)

        canvas_menu = tk.Canvas(root_menu, width = 150, height = 125)
        canvas_menu.config(bg = white, bd = 0)
        canvas_menu.pack()


        but_menu1 = tk.Button(root_menu,command = showInstructions,text = 'INSTRUCTIONS',width = 14, bg = color[topic], fg = black, font = 'none 12 bold',  padx = 0, pady = 0, bd = 0)
        canvas_menu.create_window(75,15, window = but_menu1)

        but_menu2 = tk.Button(root_menu,command = showExplore,text = 'HOME',width = 14, bg = color[topic-1], fg = black, font = 'none 12 bold',  padx = 0, pady = 0, bd = 0)
        canvas_menu.create_window(75,45, window = but_menu2)

        but_menu3 = tk.Button(root_menu,command = showAbout,text = 'ABOUT',width = 14, bg = color[topic-2], fg = black, font = 'none 12 bold',  padx = 0, pady = 0, bd = 0)
        canvas_menu.create_window(75,75, window = but_menu3)

        but_menu4 = tk.Button(root_menu,command = closeApp,text = 'EXIT',width = 14, bg = color[topic-3], fg = black, font = 'none 12 bold',  padx = 0, pady = 0, bd = 0)
        canvas_menu.create_window(75,105, window = but_menu4)
        
        root_menu.protocol("WM_DELETE_WINDOW", on_closing)
        root_menu.mainloop()
    else:
         tk.messagebox.showwarning('WARNING!',"Sorry, the menu window is already open.")
```
