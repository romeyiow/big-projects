```python
def showIntegMain():
    
    def on_closing():
        global root_menu,menuTracker,homeTracker
        
        if tk.messagebox.askokcancel("Quit", "Close differentiation window?"):
            try:
                root_menu.destroy()
            except:
                pass
            else:
                menuTracker = 0
            finally:
                tk.messagebox.showinfo("Redirecting","You'll be redirected to home.")
                homeTracker = 0
                root_integ.destroy()
                showHome()
                
    def solveInteg():
        """
        calls actual integration process then show the results
        """
        global function
        function = lambda x:eval(equation)

        myInteg(function, float(eval(ent_bLim.get())), float(eval(ent_aLim.get())), int(eval(ent_n.get()))) 
        global root_menu,menuTracker,homeTracker
        try:
            root_menu.destroy()
        except:
            pass
        else:
            menuTracker = 0
        finally:
            #tk.messagebox.showinfo("WHERE'S MENU?","Don't worry..closing the results window will redirect you to previously chosen topic.")
            showResultWin()
        
    def warningMessage():
        tk.messagebox.showwarning('INVALID INPUT!',"""Please be guided by the ff:\nFill out all entry box with appropriate inputs:\n\n
            EXPRESSION - must be in terms of 'x' only- writen in computer science way.\n\n   Ex.3*x**3 + 2*x**2 + x + 1"\n\n
            LIMITS - upper limit must be greater than or equal to lower limit.\n\n
            n - must be a COUNTING NUMBER (e.g. 1,2,3).\n\nPress "PROCEED" button to differentiate\nTIPS:\n*Use 'pi' for π. \n*Instead of 'sin**(-1)' use arcsin and so on.
            """)
        
    def warningMessage2():
        MsgBox = tk.messagebox.askquestion ('Warning...',"Its seem like your limits are not rightfully placed.Continue anyway?",icon = 'warning')
        if MsgBox == 'yes':
            return True
        else:
            return False
    def onClickInteg(event):
        """
        function that gets called whenever entry is clicked
        """
        if ent_equation.get() == ' f(x) = _______':
            ent_equation.delete(0, "end") # delete all the text in the entry
            ent_equation.insert(0, '') #Insert blank for user input
            ent_equation.config(fg = black)

    def onOutInteg(event):
        """
        function that gets called whenever entry is clicked
        """
        if ent_equation.get() == '':
            ent_equation.insert(0, ' f(x) = _______')
            ent_equation.config(fg = black)
            
    def checkInputs():
        try:
            x = 1-pi
            pos_func = eval(ent_equation.get())
            pos_b = float(eval(ent_bLim.get()))
            pos_a = float(eval(ent_aLim.get()))
            pos_n = int(eval(ent_n.get()))
            
                    
                
        except:
            warningMessage()
        else:
            global equation
            equation = ent_equation.get()
            global inputs_integ
            if pos_b < pos_a:
                if warningMessage2():
                    inputs_integ[0] = equation
                    inputs_integ[1] = ent_aLim.get()
                    inputs_integ[2] = ent_bLim.get()
                    inputs_integ[3] = ent_n.get()
                    solveInteg()
                else:
                    return
            else:
                inputs_integ[0] = equation
                inputs_integ[1] = ent_bLim.get()
                inputs_integ[2] = ent_aLim.get()
                inputs_integ[3] = ent_n.get()
                solveInteg()
            
         
    global root_integ
    root_integ = tk.Tk()
    root_integ.geometry("400x300+600+200")
    root_integ.title('NUMVERSE: INTEGRATION')
    root_integ.maxsize( width = 400, height = 300 )
    root_integ.minsize( width = 400, height = 300 )
    root_integ.resizable(0,0)
    


    #------- INTEGRATION HEADER----------
    canvas_integ_bar = tk.Canvas(root_integ, width = 400, height = 50)
    canvas_integ_bar.config(bg = color[3] , bd = 0 ,highlightthickness = 1)
    canvas_integ_bar.pack()
    
        #adds the MENU BUTTON
    but_menu_integ = tk.Button(root_integ,command = showMenu,text = 'Ξ',width = 3,bg = black, fg = color[3], font = 'none 20 bold ',  padx = 0, pady = 0, bd = 0)
    canvas_integ_bar.create_window(27,26, window = but_menu_integ)
    
        #adds the "INTEGRATION"- LABEL
    label = tk.Label(root_integ,text = "I N T E G R A T I O N", font = 'none 20 ', bg = color[3], fg = black, bd = 0) 
    canvas_integ_bar.create_window(200, 27, window = label)
    
    #------- INTEGRATION BODY----------
    canvas_integ = tk.Canvas(root_integ, width = 400, height = 250)
    canvas_integ.config(bg = black , bd = 0 ,highlightthickness = 1)
    canvas_integ.pack()
    
        #adds Integration Symbol LABEL
    lab_integSign = tk.Label(root_integ,text = "∫", font = ('Yu Gothic Light',80), bg = black , fg = color[3], bd = 0,padx = 0,pady =0) 
    canvas_integ.create_window(95, 80, window = lab_integSign)
    
        #add ENTRY box for upper limit (b)
    ent_bLim = tk.Entry(root_integ,width = 3, font = 'none 20', fg = black,highlightthickness = 0, bd = 0)
    ent_bLim.insert(0,inputs_integ[1])
    canvas_integ.create_window(150,30,window = ent_bLim)
    
        #add ENTRY box for lower limit (a)
    ent_aLim = tk.Entry(root_integ,width = 3, font = 'none 20', fg = black,highlightthickness = 0, bd = 0)
    ent_aLim.insert(0,inputs_integ[2])
    canvas_integ.create_window(40,120,window = ent_aLim)
    
        #add entry box for equation 
    ent_equation = tk.Entry(root_integ,width = 14, font = 'none 20', fg = black ,highlightthickness = 0, bd = 0)
    ent_equation.insert(0, inputs_integ[0])
    if inputs_integ[0] == ' f(x) = _______':
        ent_equation.config(fg = color[3])
    else:
        ent_equation.config(fg = black)
    ent_equation.bind('<FocusIn>', onClickInteg)
    ent_equation.bind('<FocusOut>', onOutInteg)
    canvas_integ.create_window(225,80,window = ent_equation)#x+=15
        
        #adds Integration Symbol LABEL
    lab_dxSign = tk.Label(root_integ,text = "dx", font = ('none',20,'bold'), bg = '#2a363b' , fg = color[3], bd = 0,padx = 1,pady =0) 
    canvas_integ.create_window(350, 80, window = lab_dxSign)
        
        #adds N- LABEL
    lab_dxSign = tk.Label(root_integ,text = "NUMBER OF DIVISIONS (n):", font = ('none',15,'bold'), bg = '#2a363b' , fg = color[3], bd = 0,padx = 1,pady =0) 
    canvas_integ.create_window(170, 170, window = lab_dxSign)
    
        #add ENTRY box for n
    ent_n = tk.Entry(root_integ,width = 3, font = 'none 20', fg = black,highlightthickness = 0, bd = 0)
    ent_n.insert(0,inputs_integ[3])
    canvas_integ.create_window(330,170,window = ent_n)
    
        #adds the GO BUTTON
    but_go = tk.Button(root_integ,command = checkInputs,text = 'PROCEED',width = 10, bg = color[3], fg= black, font = 'none 15 bold',  padx = 2, pady = 0, bd = 1)
    canvas_integ.create_window(200,220, window = but_go)
    
    root_integ.protocol("WM_DELETE_WINDOW", on_closing) 
    root_integ.mainloop()
    
#showIntegMain()
```
