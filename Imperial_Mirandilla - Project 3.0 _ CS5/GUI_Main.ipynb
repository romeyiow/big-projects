```python
try:
    global rtMain,method
    method = 1

    rtMain = Tk()
    rtMain.geometry("1020x750+300+20")
    rtMain.title('')
    rtMain.config(bg = clr[2])
    rtMain.resizable(0,1)    

    #----------THE HEADER-------
    def showAbout():
        messagebox.showinfo("Sorting Meers Hue","""\nINSTRUCTIONS:\n\n    (+) Learn about basic sorting methods/ algorithms by observing how they works,this time, not on array of numbers but on colors.\n        \n    (+) Press the "SORT" button to start sorting.\n \n        Note: You will be restricted to explore on other method once you click the button "SORT".\n        \n    (+) Adjust the sorting speed and color range to what pleases you even while sorting.\n    \n    (+) Expected outcome will be exactly the same as the color line where the instruction buton is connected.\n    \n    (+) HAVE SOME FUN,by guessing which color will be on its rightful place next.\n    \nABOUT:\n        This simple app was developed, by JEROME IMPERIAL (12- PHYSICS) & KEIRON MIRANDILLA (12-BIOLOGY), students of Philippine Science High School - Bicol Region Campus, in fulfilment of their 3rd Quarter Project in CS 5 Elective. It aims to help learners understand the three most common sorting algorithms (Bubble Sort, Selection Sort, & Quicksort) by simulating how it works on colors. \n\n""")
     
    canvas_header =tk.Canvas(rtMain,width = 1020, height = 80)
    canvas_header.config(bg = clr[0],highlightthickness = 0)
    canvas_header.pack()

        #-----App Name---
    txt_appName = Text(canvas_header, font = hdr, bg = clr[0],width = 15,height = 1, fg = clr[6],highlightthicknes = 0 ,bd = 0,spacing3 = 0,spacing2 = 0,wrap = WORD,pady = 10,padx = 5)
    txt_appName.insert('1.15','SortingMeetsHue')#txt_code.tag_configure("left", justify='left')
    txt_appName.tag_add("Hue","1.12","1.15")
    txt_appName.tag_config("Hue",foreground = clr[2])
    txt_appName.config(state = DISABLED)
    canvas_header.create_window(160, 50, window = txt_appName)

   
    txt_name = Text(canvas_header, font = ntxt, bg = clr[0],width = 45,height = 1, fg = clr[5],highlightthicknes = 0 ,bd = 0,spacing1 = 0,spacing3 = 0,spacing2 = 0,wrap = CHAR,pady = 5,padx = 5)
    txt_name.insert(INSERT,'I M P E R I A L   &  M I R A N D I L L A   ©   2 0 2 0  ')#txt_code.tag_configure("left", justify='left')
    txt_name.config(state = DISABLED)
    canvas_header.create_window(860, 60, window = txt_name)



    #--------CREATE TABS----------
    canvas_tabs = Canvas(rtMain, width = 1020, height = 60)
    canvas_tabs.config(bg = clr[2], bd = 0,highlightthickness = 0 )
    canvas_tabs.pack()


        #----ADDS BUTTONS
    global but_bubble, but_quick, but_selection
    but_bubble = Button(canvas_tabs,command = showBubble, text = 'B U B B L E   S O R T')
    but_bubble.config (width = 27,height = 3,bg = clr[5], disabledforeground = clr[0], font = ttxt,relief = FLAT,state = DISABLED, bd = 1,highlightthickness = 1,highlightbackground = clr[0],  padx = 0, pady = 0)
    canvas_tabs.create_window(160, 30, window = but_bubble)

    but_quick = Button(canvas_tabs,command = showQuick, text = 'Q U I C K S O R T')
    but_quick.config(width = 27,height = 3, bg = clr[2], fg = clr[0], disabledforeground = clr[1], font = ttxt,state = NORMAL, bd = 1,highlightthickness = 1,highlightbackground = clr[0],  padx = 0, pady = 0)
    canvas_tabs.create_window(510, 30, window = but_quick)


    but_selection = Button(canvas_tabs,command = showSelection, text = 'S E L E C T I O N   S O R T')
    but_selection.config(width = 27,height = 3, bg = clr[2], disabledforeground = clr[1], fg = clr[0], font = ttxt,state = NORMAL, bd = 1,highlightthickness = 1,highlightbackground = clr[0],  padx = 0, pady = 0)
    canvas_tabs.create_window(860, 30, window = but_selection)



    #--------CREATE INFO CONTAINER----------
    global canvas_info, txt_info,lab_perfW2,lab_perfB2,lab_perfA2

    canvas_info = Canvas(rtMain, width = 1020, height = 330)
    canvas_info.config(bg = clr[0], bd = 0,highlightthickness = 0 ,closeenough = 1,confine = True )
    canvas_info.pack()


        ##1\2------LABEL: "What gives"------------
    lab_info = Label(canvas_info,text = "WHAT GIVES?",font = hdr1, bg = clr[0], fg = clr[4] , bd = 0) 
    canvas_info.create_window(240, 50, window = lab_info)

        ##2\2----TEXT: Description for Bubble Sort

    txt_info = tk.scrolledtext.ScrolledText(canvas_info,font = ntxt1,width = 47,yscrollcommand = True,height = 5, bg = clr[0], fg = clr[6],spacing3 = 3,spacing2 = 1 ,bd = 0,wrap = WORD)
    txt_info.insert(INSERT,des[method])
    txt_info.tag_configure("left", justify='left')
    txt_info.config(state = DISABLED)
    canvas_info.create_window(250, 130, window = txt_info)


        ##1\4------LABEL: "Performances"------------
    lab_perf = Label(canvas_info,text = "PERFORMANCE",font = hdr1, bg = clr[0], fg = clr[4] , bd = 0) 
    canvas_info.create_window(250, 230, window = lab_perf)

        ##2.1\4------LABEL: "Worse-CasePerformances "------------
    lab_perfW1 = Label(canvas_info,text = "WORSE-CASE:",font = ntxt1, bg = clr[0], fg = clr[4] , bd = 0) 
    canvas_info.create_window(90, 260, window = lab_perfW1)

        ##2.2\4------LABEL: "Worse-CasePerformances "------------
    lab_perfW2 = Label(canvas_info,text = " O(n^2) comparisons & O(n^2) swaps",font = ntxt2, bg = clr[0], fg = clr[2] , bd = 0) 
    canvas_info.create_window(275, 260, window = lab_perfW2,tags = 'worsePerf' )



        ##3.1\4------LABEL: "Best-CasePerformances "------------
    lab_perfB1 = Label(canvas_info,text = "BEST-CASE: ",font = ntxt1, bg = clr[0], fg = clr[4] , bd = 0) 
    canvas_info.create_window(80, 285, window = lab_perfB1)

        ##3.2\4------LABEL: "Best-CasePerformances "------------
    lab_perfB2 = Label(canvas_info,text = " O(n) comparisons & O(1) swaps",font = ntxt2, bg = clr[0], fg = clr[2] , bd = 0) 
    canvas_info.create_window(235, 285, window = lab_perfB2,tags = 'bestPerf' )

        ##4.1\4------LABEL: "Best-CasePerformances "------------
    lab_perfA1 = Label(canvas_info,text = "AVERAGE-CASE:",font = ntxt1, bg = clr[0], fg = clr[4] , bd = 0) 
    canvas_info.create_window(95, 310, window = lab_perfA1)

         ##4.2\4------LABEL: "Best-CasePerformances "------------
    lab_perfA2 = Label(canvas_info,text = " O(n^2) comparisons & O(n^2) swaps",font = ntxt2, bg = clr[0], fg = clr[2] , bd = 0) 
    canvas_info.create_window(290, 310, window = lab_perfA2,tags = 'avgPerf')



        ##------ALGORITHM CONTAINER---------
    global canvas_algo, txt_code
    canvas_algo = Canvas(canvas_info, width = 510, height = 300)
    canvas_algo.config(bg = clr[1], bd = 3,highlightthickness = 2 )
    canvas_info.create_window(750,165,window = canvas_algo)

            ###------LABEL: ALGORITHM-------------
    lab_code = Label(canvas_algo,text = "GENERAL ALGORITHM:",font = hdr2, bg = clr[1], fg = clr[6] , bd = 0) 
    canvas_algo.create_window(130, 40, window = lab_code)
            ##------BUTTON: INFO


            ###-----TEXT BOX: ALGORITHM--------- ,Webdings
    code = inspect.getsource(bubbleSort)
    txt_code = tk.scrolledtext.ScrolledText(canvas_algo, font = cd, width = 60,height = 11, bg = clr[5], fg = clr[0] ,bd = 0,spacing3 = 0,spacing2 = 0,wrap = WORD,pady = 10,padx = 5)
    txt_code.insert(INSERT,code)#txt_code.tag_configure("left", justify='left')
    txt_code.tag_add("def","1.0","1.3")
    txt_code.tag_add("funcName","1.4","1.14")
    txt_code.tag_add("return","8.4","8.10")
    txt_code.tag_config("def",foreground = clr[-1])
    txt_code.tag_config("funcName",foreground = clr[-2])
    txt_code.tag_config("return",foreground = clr[-1])
    txt_code.config(state = DISABLED)
    canvas_algo.create_window(260, 185, window = txt_code)



    #--------CODES FOR SORTING---------- 

    def showGuideGradient():
        for i in range(50):
            exec('global canvas_{0}'.format(i))
            exec('canvas_{0} = Canvas(canvas_sorting, width = 40, height = 15)'.format(i))
            exec('canvas_{0}.config(bg = c[{0}], bd = 0,highlightthickness = 0)'.format(i))
            exec('canvas_sorting.create_window(20+({0}*20),7.5,window = canvas_{0})'.format(i))

    def showUnsorted():
        global c,x,but_sort,but_shuffle
        x = random.sample(range(50),50)
        for i in range(50):
            canvas_sorting.create_rectangle(10+i*20, 150, 30+i*20, 200, fill = c[x[i]], outline =c[x[i]], tags = 'rect_{}'.format(x[i]))


    def shuffle():

        global but_sort,but_shuffle,lab_action
        but_sort.config(state = DISABLED, relief = SUNKEN, fg = clr[-3],bg = clr[0])
        for i in range(50):
            canvas_sorting.delete("rect_{}".format(i))
        showUnsorted()
        
        lab_action.config(text = ' “CLICK SORT BUTTON” ')
        but_sort.config(bg = clr[2], fg = clr[0],state = NORMAL, relief = RAISED)

    def updateGradient():
        global cStart,c,cEnd,x
        cc = ['black','red','pink','orange','yellow','green','cyan','blue','violet','white']
        cStart = Color(Svar_choice.get())
        cEnd =  Color(Evar_choice.get())
        if Svar_choice.get() == Evar_choice.get():
            messagebox.showwarning("WARNING!","Color range must not start and end with same the color. Gradient effect will not be visible.We will now randomize your gradient.")
            e = cc.index(Evar_choice.get())
            cStart = Color(cc[e-3])
            Svar_choice.set(cc[e-3])
        c = list(cStart.range_to(cEnd,50))
        for i in range(50):
            exec('global canvas_{0}'.format(i))
            exec('canvas_{0} = Canvas(canvas_sorting, width = 40, height = 15)'.format(i))
            exec('canvas_{0}.config(bg = c[{0}], bd = 0,highlightthickness = 0)'.format(i))
            exec('canvas_sorting.create_window(20+({0}*20),7.5,window = canvas_{0})'.format(i))

        for i in range(50):
            canvas_sorting.itemconfigure('rect_{}'.format(x[i]),fill = c[x[i]], outline =c[x[i]])

    def change_dropdown(*args):# on change dropdown value
        updateGradient()

    def updateSpeed(event):
        global speed 
        speed =  scl_speed.get()

    def callSort():
        global x,but_sort,but_shuffle,method

        but_shuffle.config(state = DISABLED)
        but_sort.config(state = DISABLED, relief = SUNKEN, fg = clr[-3],bg = clr[0])

        if method == 1:
            newbubble(x)
        elif method == 2:
            newquickSortHelper(x,0,len(x)-1,1)
        else:
            newselectionSort(x)

        but_shuffle.config(state = NORMAL)



    """
    The following lines of codes will generate the canvas and 
        widgets that will be needed for animating the sorting process
    """
    canvas_sorting = Canvas(rtMain, width = 1020, height = 360)
    canvas_sorting.config(bg = clr[1], bd = 0,highlightthickness = 0,highlightbackground = clr[2])
    canvas_sorting.pack()
    
    but_info = Button(canvas_sorting,font = ('Webdings',15),height = 0,relief= FLAT,  text = 'i',command = showAbout, bd= 1,width = 1, bg = clr[2], fg = clr[0],state = NORMAL,highlightthickness = 0,  padx = 0,pady = 0)
    canvas_sorting.create_window(995,45,window = but_info)
    canvas_sorting.create_rectangle(980, 5, 1010, 80, fill = clr[2], outline = clr[2], tags = 'extra')
    canvas_sorting.create_rectangle(980, 70, 1010, 80, fill = clr[1], outline = clr[1], tags = 'extra')
    canvas_sorting.create_rectangle(980, 75, 1010, 80, fill = clr[6], outline = clr[6], tags = 'extra')

    showGuideGradient()
      
    global lab_action
    lab_action = Label(canvas_sorting,text = ' “CLICK SORT BUTTON” ',height =1,pady = 0,padx =0,font = ('Tw Cen MT',20), bg = clr[1], fg = clr[6] , bd = 0) 
    canvas_sorting.create_window(510, 80, window = lab_action )

    
    showUnsorted() 

    lab_colRange = Label(canvas_sorting,text = "Color Range:",font = hdr3, bg = clr[1], fg = clr[6] , bd = 0) 
    canvas_sorting.create_window(70, 260, window = lab_colRange)



    Svar_choice = tk.StringVar(canvas_sorting)
    text_choices = ['black','red','pink','orange','yellow','green','cyan','blue','violet','white']
    choices = {'black','red','pink','orange','yellow','green','cyan','blue','violet','white'}# Dictionary with options
    Svar_choice.set(text_choices[6]) # set the default option
    dd_methods = tk.OptionMenu(canvas_sorting, Svar_choice, *choices)
    dd_methods.config(width = 7, font = ntxt2, fg = clr[1],highlightthickness = 0, bd = 0,bg = clr[4])
    canvas_sorting.create_window(180,260,window = dd_methods)
    Svar_choice.trace('w', change_dropdown)

    lab_colTo = Label(canvas_sorting,text = "-",font = hdr2, bg = clr[1], fg = clr[6] , bd = 0) 
    canvas_sorting.create_window(240, 260, window = lab_colTo)

    Evar_choice = tk.StringVar(canvas_sorting)
    text_choices = ['black','red','pink','orange','yellow','green','cyan','blue','violet','white']
    choices = {'black','red','pink','orange','yellow','green','cyan','blue','violet','white'}# Dictionary with options
    Evar_choice.set(text_choices[2]) # set the default option
    dd_methods = tk.OptionMenu(canvas_sorting, Evar_choice, *choices)
    dd_methods.config(width = 7, font = ntxt2, fg = clr[1],highlightthickness = 0, bd = 0,bg = clr[4])
    canvas_sorting.create_window(300,260,window = dd_methods)
    Evar_choice.trace('w', change_dropdown)


    lab_colTo = Label(canvas_sorting,text = "Sorting Speed:",font = hdr3, bg = clr[1], fg = clr[6] , bd = 0) 
    canvas_sorting.create_window(430, 260, window = lab_colTo)


    scl_speed = Scale(canvas_sorting,showvalue = 0,command = updateSpeed,font = ntxt1,from_= 1, to = 10,orient = HORIZONTAL,sliderlength = 20,width = 5, length = 150) 
    canvas_sorting.create_window(580, 260, window = scl_speed)
    scl_speed.set(speed)
    #scl_speed.bind('<FocusOut>',updateSpeed)

    global but_sort,but_shuffle
    but_sort = Button(canvas_sorting, disabledforeground = clr[1],font = hdr3,relief = RAISED,  text = 'SORT',command = callSort, bd = 1,width = 13,height = 1, bg = clr[2], fg = clr[0],state = NORMAL,highlightthickness = 1,highlightbackground = clr[0],  padx = 0, pady = 0)
    canvas_sorting.create_window(780,260,window = but_sort)

    but_shuffle = Button(canvas_sorting, disabledforeground = clr[1],font =  hdr3,relief = RAISED,  text = 'SHUFFLE',command = shuffle, bd= 1,width = 13,height = 1, bg = clr[2], fg = clr[0],state = NORMAL,highlightthickness = 1,highlightbackground = clr[0],  padx = 0, pady = 0)
    canvas_sorting.create_window(935,260,window = but_shuffle)

    rtMain.mainloop()
except:
    print('PLEASE RE-RUN THE PROGRAM/APP')
```
