### Animation

```python
def updateOrder(n,jx,kx,aList):
    global speed 
    
    jx1, jy1, jx2, jy2 = canvas_sorting.bbox("rect_{}".format(aList[n]))
    kx1, ky1, kx2, kcury2 = canvas_sorting.bbox("rect_{}".format(aList[n+1]))
   
    if jx1 == kx:
        pass 
    else:
        canvas_sorting.move('rect_{}'.format(aList[n]),-speed/20,0)
    if kx1 == jx:
        pass
    else:
        canvas_sorting.move('rect_{}'.format(aList[n+1]),speed/20,0)
   
    if jx1 == kx and kx1 == jx:
        rtMain.update()
        return
    rtMain.update()
    canvas_sorting.after(0, updateOrder(n,jx,kx,aList))
    
def prepare_ToCompare(n,aList,j,k):
    global speed 
    jx1, jy1, jx2, jy2 = canvas_sorting.bbox("rect_{}".format(aList[n]))
    kx1, ky1, kx2, ky2 = canvas_sorting.bbox("rect_{}".format(aList[n+1]))
    if jy1 == 125:
        pass
    else:
        canvas_sorting.move('rect_{}'.format(aList[n]),0,-speed/10)
    if ky1 == 125:
        pass
    else:
        canvas_sorting.move('rect_{}'.format(aList[n+1]),0,-speed/10)
    if jy1 == 125 and ky1 == 125:
        rtMain.update()
        return
    
    rtMain.update()
    canvas_sorting.after(0,prepare_ToCompare(n,aList,j,k))

def reset_ToCompare(n,aList,j,k):
    global speed 
    jx1, jy1, jx2, jy2 = canvas_sorting.bbox("rect_{}".format(aList[n]))
    kx1, ky1, kx2, ky2 = canvas_sorting.bbox("rect_{}".format(aList[n+1]))
   
    if jy1 == 150:
        pass
    else:
        canvas_sorting.move('rect_{}'.format(aList[n]),0,speed/10)
    if ky1 == 150:
        pass
    else:
        canvas_sorting.move('rect_{}'.format(aList[n+1]),0,speed/10)
    if jy1 == 150 and jy1 == 150:
        return
    rtMain.update()
    canvas_sorting.after(0,reset_ToCompare(n,aList,j,k))
    
def newbubble (aList):
    global but_bubble, but_quick, but_selection,lab_action
    but_bubble.config (state = DISABLED)
    but_quick.config(state = DISABLED)
    but_selection.config(state = DISABLED)
    
    exchanges = True
    passnum = len(aList) - 1
    while passnum > 0 and exchanges:
        exchanges = False
        for i in range (passnum):
            if i%3 == 0: 
                lab_action.config(text = ' “COMPARE...” ')
            elif i%3 == 2:
                lab_action.config(text = ' “COMPARE..” ')
            else:
                lab_action.config(text = ' “COMPARE.” ')
            prepare_ToCompare(i,aList,-0.5,-0.5)
            if aList[i] > aList[i+1]:
                exchanges = True
                temp = aList[i]
                aList[i] = aList[i+1]
                aList[i+1] = temp
                jx1, jy1, jx2, jy2 = canvas_sorting.bbox('rect_{0}'.format(aList[i]))
                kx1, ky1, kx2, ky2 = canvas_sorting.bbox('rect_{0}'.format(aList[i+1]))
                reset_ToCompare(i,aList,0.5,0.5)
                rtMain.update()
                lab_action.config(text = ' “SWAP” ')
                updateOrder(i,jx1,kx1,aList)
            else:
                rtMain.update()
                reset_ToCompare(i,aList,0.5,0.5)
        rtMain.update()
        passnum -= 1    
    
    lab_action.config(text = ' “CLICK SHUFFLE BUTTON” ')
    but_quick.config(state = NORMAL)
    but_selection.config(state = NORMAL)
```
