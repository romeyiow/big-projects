### ANIMATION

```python
def resetBasis(prev,new,aList,ylim,speed):
    if prev == new:
        liftBox(prev,aList,ylim,speed)        
    else:
        returnBox(prev,aList,ylim,speed)
        liftBox(new,aList,100,speed)
        
def returnBox(loc,aList,ylim,speed):
    jx1, jy1, jx2, jy2 = canvas_sorting.bbox("rect_{}".format(aList[loc]))
    
    if jy1 == 150:
        return
    else:
        canvas_sorting.move('rect_{}'.format(aList[loc]),0,1)
    rtMain.update()
    canvas_sorting.after(0,returnBox(loc,aList,ylim,speed))       

def returnBothBox(posMax,loc,aList,ylim,speed):
    jx1, jy1, jx2, jy2 = canvas_sorting.bbox("rect_{}".format(aList[posMax]))
    kx1, ky1, kx2, ky2 = canvas_sorting.bbox("rect_{}".format(aList[loc]))
    speed = 1
    if jy1 == 150:
        pass
    else:
        canvas_sorting.move('rect_{}'.format(aList[posMax]),0,speed)
    if ky1 == 150:
        pass
    else:
        canvas_sorting.move('rect_{}'.format(aList[loc]),0,-speed)
    if jy1 == 150 and ky1 == 150:
        return
    rtMain.update()
    canvas_sorting.after(0,returnBothBox(posMax,loc,aList,ylim,speed)) 
    
def liftBox(loc,aList,ylim,speed):
    jx1, jy1, jx2, jy2 = canvas_sorting.bbox("rect_{}".format(aList[loc]))
  
    corrector = 1
    if jy1 < ylim:
        corrector = -1
    if jy1 == ylim:
        return
    else:
        canvas_sorting.move('rect_{}'.format(aList[loc]),0,-speed/10*corrector)
    rtMain.update()
    canvas_sorting.after(0,liftBox(loc,aList,ylim,speed))
    
def dropBox(loc,aList,ylim,speed):
    jx1, jy1, jx2, jy2 = canvas_sorting.bbox("rect_{}".format(aList[loc]))
    
    if jy1 == ylim:
        return
    else:
        canvas_sorting.move('rect_{}'.format(aList[loc]),0,speed/10)
    rtMain.update()
    canvas_sorting.after(0,dropBox(loc,aList,ylim,speed))  
    
def settleMax(posMax,lim,plim,xlim,aList,speed):
    if posMax != lim:
        dropBox(lim,aList,200,speed)
        jx1, jy1, jx2, jy2 = canvas_sorting.bbox("rect_{}".format(aList[posMax]))
        kx1, ky1, kx2, ky2 = canvas_sorting.bbox("rect_{}".format(aList[lim]))
        
        inc = 0
    
        if speed <= 3 and (xlim - jx1) > 0:
            inc = ((xlim - jx1)//300)

        if (xlim - jx1)//20 < 15:
            speed = 1
            if (xlim - jx1)//20 < 5:
                speed = 0.5
        if (xlim - jx1)//20 > 15 and (xlim - jx1)//20 < 35 and speed > 5:
            speed = 3
            
        if (speed + inc) >= 10:
            inc = 0
            
        if jx1 == xlim:
            pass
        else:
            canvas_sorting.move('rect_{}'.format(aList[posMax]),(speed+inc),0)
        if kx1 == plim:
            pass
        else:
            canvas_sorting.move('rect_{}'.format(aList[lim]),-(speed+inc),0)
        if jx1 == xlim and  kx1 == plim:
            returnBothBox(posMax,lim,aList,150,speed)
            return
        rtMain.update()
        
        canvas_sorting.after(0,settleMax(posMax,lim,plim,xlim,aList,speed))
    else:
        returnBox(posMax,aList,150,speed)
        
def newselectionSort (aList):
    global but_bubble, but_quick, but_selection,lab_action
    
    
    but_bubble.config (state = DISABLED)
    but_quick.config(state = DISABLED)
    but_selection.config(state = DISABLED)
   
    for fillslot in range (len(aList) - 1, 0, -1):
        global speed
        lab_action.config(text = ' “INITIALIZE HIGHEST VALUE” ')
        positionOfMax = 0
        resetBasis(positionOfMax,0,aList,100,speed)
        
        for location in range (1, fillslot + 1):
            lab_action.config(text = ' “PASS AND COMPARE” ')
            liftBox(location,aList,125,speed)
    
            if aList [location] > aList [positionOfMax]:
                lab_action.config(text = ' “UPDATE HIGHEST VALUE” ')
                resetBasis(positionOfMax,location,aList,125,speed)
                positionOfMax = location
            else:
                returnBox(location,aList,125,speed)
                
        temp = aList [fillslot]
        jx1, jy1, jx2, jy2 = canvas_sorting.bbox("rect_{}".format(aList[positionOfMax]))
        kx1, ky1, kx2, ky2 = canvas_sorting.bbox("rect_{}".format(aList[fillslot]))
        lab_action.config(text = ' “PLACE HIGHEST VALUE IN RIGHT LOCATION” ')
        settleMax(positionOfMax,fillslot,jx1,kx1,aList,speed)
        aList [fillslot] = aList [positionOfMax]
        aList[positionOfMax] = temp
     
    but_bubble.config(state = NORMAL)
    but_quick.config(state = NORMAL)
```
