### ANIMATION

```python
def liftMark(loc,direction,goal,aList,speed):
    
    if abs(loc) < len(aList):
        jx1, jy1, jx2, jy2 = canvas_sorting.bbox("rect_{}".format(aList[loc]))
        
        if jy1 == goal:
            return
        else:
            canvas_sorting.move('rect_{}'.format(aList[loc]),0,speed/10*direction)

        rtMain.update()
        canvas_sorting.after(0,liftMark(loc,direction,goal,aList,speed))  
    else:
        return
def settleMarks(rightmark,leftmark,aList,speed):
    
    if rightmark != leftmark:
        lx1, ly1, lx2, ly2 = canvas_sorting.bbox("rect_{}".format(aList[leftmark])) 
        rx1, ry1, rx2, ry2 = canvas_sorting.bbox("rect_{}".format(aList[rightmark])) 

        if ly1 == 125:
            rtMain.update()
            pass
        else:
            canvas_sorting.move('rect_{}'.format(aList[leftmark]),0,-speed/10)
        if ry1 == 175:
            rtMain.update()
            pass
        else:
            canvas_sorting.move('rect_{}'.format(aList[rightmark]),0,speed/10)
        if ly1 == 125 and ry1 == 175:
            rtMain.update()
            return
        rtMain.update()
        canvas_sorting.after(0,settleMarks(rightmark,leftmark,aList,speed))
    else:
        return
    
def swapMarks(leftmark,rightmark,Llim,Rlim,aList,speed):

    lx1, ly1, lx2, ly2 = canvas_sorting.bbox("rect_{}".format(aList[leftmark])) 
    rx1, ry1, rx2, ry2 = canvas_sorting.bbox("rect_{}".format(aList[rightmark])) 
    inc = 0
    
    if speed <= 3 and (Rlim - lx1) > 0:
        inc = ((Rlim - lx1)//300)
        
    if (Rlim - lx1)//20 < 10:
        speed = 1
        if (Rlim - lx1)//20 < 5:
            speed = 0.5

    if (speed + inc) >= 10:
        inc = 0
    
    if lx1 == Rlim:
        pass
    else:
        canvas_sorting.move('rect_{}'.format(aList[leftmark]),speed+inc,0)
        
    if rx1 == Llim:
        pass
    else:
        canvas_sorting.move('rect_{}'.format(aList[rightmark]),-(speed+inc),0)
        
    if lx1 == Rlim and  rx1 == Llim: 
        return
    
    rtMain.update()
    canvas_sorting.after(0,swapMarks(leftmark,rightmark,Llim,Rlim,aList,speed))

def swapRightAndPivot(start,rightmark,Slim,Rlim,aList,speed):
   
    sx1, sy1, sx2, sy2 = canvas_sorting.bbox("rect_{}".format(aList[start])) 
    rx1, ry1, rx2, ry2 = canvas_sorting.bbox("rect_{}".format(aList[rightmark]))  
    
    inc = 0
    if speed <= 3 and (Rlim - sx1) > 0:
        inc = ((Rlim - sx1)//300)
        
    if (Rlim - sx1)//20 < 10:
        speed = 1
        if (Rlim - sx1)//20 < 5:
            speed = 0.5

    if (speed + inc) >= 10:
        inc = 0
        
    if sx1 == Rlim:
        pass
    else:
        canvas_sorting.move('rect_{}'.format(aList[start]),speed+inc,0)
    
    if rx1 == Slim:
        pass
    
    else:
        canvas_sorting.move('rect_{}'.format(aList[rightmark]),-(speed+inc),0)
    
    if sx1 == Rlim and  rx1 == Slim:
        return
    
    rtMain.update()
    canvas_sorting.after(0,swapRightAndPivot(start,rightmark,Slim,Rlim,aList,speed))

    
def resetAll(aList):
    
    for loc in range (len(aList)):
        jx1, jy1, jx2, jy2 = canvas_sorting.bbox("rect_{}".format(aList[loc]))
        if jy1 < 150: #(loc,direction,goal,aList)
            liftMark(loc,1,150,aList,speed)
        elif jy1 > 150:
            liftMark(loc,-1,150,aList,speed)            
    return
def newquickSortHelper(aList,start,end,pas):
    global but_bubble, but_quick, but_selection,lab_action
    but_bubble.config (state = DISABLED)
    but_quick.config(state = DISABLED)
    but_selection.config(state = DISABLED)
   
    newquickSort(aList,start,end,pas)
    lab_action.config(text = ' “PRESS SHUFFLE BUTTON” ')
    
    but_bubble.config(state = NORMAL)
    but_selection.config(state = NORMAL)
    
def newquickSort(aList,start,end,pas):
    
    global speed,lab_action
    pivotvalue = aList [start]

    leftmark = start + 1 
    rightmark = end
  
    done = False

    #liftMark(leftmark,1,175,aList,speed)
    #liftMark(rightmark,-1,125,aList,speed)
    
    while not done:
        lab_action.config(text = ' “LEFTMARK MOVING” ')
        while leftmark <= rightmark and aList[leftmark] <=  pivotvalue:
            if leftmark < rightmark:
                liftMark(leftmark,1,175,aList,speed)
            leftmark = leftmark + 1
        lab_action.config(text = ' “LEFTMARK STOPS” ')
        rtMain.update()
        if speed < 4:
            time.sleep(1)

        lab_action.config(text = ' “RIGHTMARK MOVING” ')
        while  rightmark >= leftmark and aList[rightmark] >=  pivotvalue:        
            if rightmark >= leftmark:
                liftMark(rightmark,-1,125,aList,speed)
            elif rightmark < leftmark:
                liftMark(rightmark,-1,150,aList,speed)
            rightmark = rightmark - 1
        lab_action.config(text = ' “RIGHTMARK STOPS” ')
        rtMain.update()
        if speed < 4:
            time.sleep(1)
        
        if rightmark < leftmark:

            sx1, sy1, sx2, sy2 = canvas_sorting.bbox("rect_{}".format(aList[start])) 
            rx1, ry1, rx2, ry2 = canvas_sorting.bbox("rect_{}".format(aList[rightmark])) 
            
            lab_action.config(text = ' “MARKS CROSSED.SWAP RIGHMARK AND PIVOT VALUE” ')
            swapRightAndPivot(start,rightmark,sx1,rx1,aList,speed)
            done = True
                
        else: 
            lx1, ly1, lx2, ly2 = canvas_sorting.bbox("rect_{}".format(aList[leftmark])) 
            rx1, ry1, rx2, ry2 = canvas_sorting.bbox("rect_{}".format(aList[rightmark])) 
            
            lab_action.config(text = ' “SWAP RIGHMARK AND LEFTMARK” ')
            settleMarks(rightmark,leftmark,aList,speed)
            swapMarks(leftmark,rightmark,lx1,rx1,aList,speed)
           
            
            temp = aList [leftmark]
            aList [leftmark] = aList[rightmark]
            aList [rightmark] = temp

    
    
    temp = aList [start]
    aList[start] = aList[rightmark]
    aList[rightmark] = temp

    leftSide = aList[start:rightmark]
    rightSide = aList[rightmark+1:end+1]

    if len(leftSide) >= 2:
        lab_action.config(text = ' “PREPARING ITEMS FOR QUICKSORT OF LEFTSIDE” ')
        resetAll(leftSide)
        leftSide = newquickSort(leftSide,0,len(leftSide)-1,pas+1)

    if len(rightSide) >= 2:
        lab_action.config(text = ' “PREPARING ITEMS FOR QUICKSORT OF RIGHTSIDE” ')
        resetAll(rightSide)
        rightSide = newquickSort(rightSide,0,len(rightSide)-1,pas+1)
        
    resetAll(leftSide + [pivotvalue] + rightSide)
    return leftSide + [pivotvalue] + rightSide
```
